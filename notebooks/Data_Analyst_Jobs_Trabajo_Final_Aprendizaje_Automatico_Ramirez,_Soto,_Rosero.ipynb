```python
data_analyst_jobs=pd.concat([data_africa,data_usa,data_canada],axis=0).reset_index()

import pandas as pd
from functools import reduce
import re
from ast import literal_eval as evalu

# Concatenar DataFrames
data_analyst_jobs = pd.concat([data_africa, data_usa, data_canada], axis=0).reset_index(drop=True)

# Funci√≥n para limpiar el salario
def clean_sal(text):
    if isinstance(text, str):
        list_nums = re.findall(r"[-+]?\d*\.*\d+", text)
        return float(reduce(lambda a, b: (a + b) / 2, map(float, list_nums)))
    return None

# Aplicar la funci√≥n clean_sal
data_analyst_jobs["sal"] = data_analyst_jobs["salary"].apply(clean_sal)
data_analyst_jobs["sal"].fillna(data_analyst_jobs["sal"].median(), inplace=True)
data_analyst_jobs.drop(columns=["salary"], inplace=True)

# Establecer el √≠ndice
if 'index' in data_analyst_jobs.columns:
    data_analyst_jobs.set_index("index", inplace=True)

# Funci√≥n para formar criterios
def form_criteria(string):
    try:
        string = evalu(string)
        if len(string) > 0:
            return reduce(lambda a, b: {**a, **b}, string)
    except (ValueError, SyntaxError):
        return {}
    return {}

# Aplicar la funci√≥n form_criteria
criteria_df = data_analyst_jobs["criteria"].apply(form_criteria).apply(pd.Series)
data_analyst_jobs_temp = pd.concat([data_analyst_jobs.drop(["criteria"], axis=1), criteria_df], axis=1).reset_index(drop=True)

# Eliminar columnas innecesarias
data_analyst_jobs_temp.drop(columns=data_analyst_jobs_temp.iloc[:, -4:].columns.tolist(), inplace=True)
data_analyst_jobs_temp["Country"] = data_analyst_jobs_temp["location"].apply(lambda x: x.split(",")[-1])


```

### Preprocesamiento

```python
# Convertir posted_date a datetime para asegurar que sea el tipo correcto
df['posted_date'] = pd.to_datetime(df['posted_date'])
```

```python
# Imprimimos los primeros 10 salarios para poder ver correctamente la estructura que manejan
df['salary'].unique()[:10]
```

```python
# Funcion para corregir el formato de los salarios.

def extract_salary(salary_string):

    # Si est√° en blanco, se convierte a tipo "nan"
    if pd.isna(salary_string):
        return np.nan, np.nan, np.nan

    # Eliminar caracteres no num√©ricos excepto el punto decimal, la coma y el guion
    salary_string = re.sub(r'[^\d.,-]', '', str(salary_string))

    # Separar los valores m√≠nimo y m√°ximo
    values = re.findall(r'\d+\.?\d*', salary_string)

    if len(values) >= 2:
        min_salary = float(values[0])
        max_salary = float(values[-1])
    elif len(values) == 1:
        min_salary = max_salary = float(values[0])
    else:
        return np.nan, np.nan, np.nan

    # Calcular el salario promedio
    avg_salary = (min_salary + max_salary) / 2

    return min_salary, max_salary, avg_salary

# Aplicar la funci√≥n de extracci√≥n de salarios
df['salary_min'], df['salary_max'], df['salary_avg'] = zip(*df['salary'].apply(extract_salary))
```

```python
# Determinar si el salario es por hora o anual
df['salary_type'] = df['salary'].apply(lambda x: 'Hourly' if pd.notna(x) and any(char.isdigit() for char in str(x)) and float(re.findall(r'\d+\.?\d*', str(x))[0]) < 1000 else 'Annual')
```

```python
# Convertir salarios por hora a anuales (asumiendo 2080 horas de trabajo al a√±o)
mask = df['salary_type'] == 'Hourly'
df.loc[mask, 'salary_min'] *= 2080
df.loc[mask, 'salary_max'] *= 2080
df.loc[mask, 'salary_avg'] *= 2080
```

```python
# Funci√≥n para extraer la industria y la funci√≥n del trabajo
def extract_criteria(criteria):
    industries = []
    job_function = []
    try:
        # Convertir la cadena en una lista de diccionarios
        criteria_list = ast.literal_eval(criteria)
        # Recorrer la lista y buscar las claves 'Industries' y 'Job function'
        for criteria_dict in criteria_list:
            if 'Industries' in criteria_dict:
                industries.extend(criteria_dict['Industries'].split(','))
            if 'Job function' in criteria_dict:
                job_function.extend(criteria_dict['Job function'].split(','))
    except (ValueError, SyntaxError):
        pass
    return industries, job_function

# Aplicar la funci√≥n y crear dos nuevas columnas en el DataFrame
df[['Industries', 'Job function']] = df['criteria'].apply(lambda x: pd.Series(extract_criteria(x)))
```
